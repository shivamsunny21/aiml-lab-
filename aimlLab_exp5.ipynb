```python
df.hist(figsize=(10, 8), bins=20, color='skyblue', edgecolor='black')
plt.suptitle('Distribution of Numerical Columns', fontsize=16)
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(data=df.select_dtypes(include=np.number))
plt.title('Boxplot of Numerical Columns', fontsize=16)
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.countplot(y='MATHS',data=df, order=df['MATHS'].value_counts().index)
plt.title('Top marks in maths', fontsize=16)
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(y='SCIENCE', data=df, order=df['SCIENCE'].value_counts().index[:10])
plt.title('Top marks in science', fontsize=16)
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(y='ENGLISH', data=df, order=df['ENGLISH'].value_counts().index[:10])
plt.title('Top marks in english', fontsize=16)
plt.show()

```
